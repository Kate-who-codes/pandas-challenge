```python
#Calculate the total number of schools
schoolData_complete["school_count"]=school_data.school_name.count()
```

```python
#Calculate the total number of students
schoolData_complete["total_students"]=schoolData_complete.student_ID.count()
```

```python
#Calculate the total budget
schoolData_complete["district_budget"]=school_data.budget.sum()
schoolData_complete["district_budget"].head()
```

```python
#Calculate the average math score
schoolData_complete["avg_math_score"]=schoolData_complete.math_score.mean()
```

```python
#Calculate the average reading score
schoolData_complete["avg_reading_score"]=schoolData_complete.reading_score.mean()
schoolData_complete.avg_reading_score.head()
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
schoolData_complete["overall_avg_score"]=(schoolData_complete.math_score.mean()+schoolData_complete.reading_score.mean())/2
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
passing_math=schoolData_complete.loc[(
    schoolData_complete["math_score"] > 70)]
passing_math.math_score.count()
```

```python
schoolData_complete["per_passing_math"]=(passing_math.math_score.count()/schoolData_complete.math_score.count())
schoolData_complete.per_passing_math.head(1)
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading=schoolData_complete.loc[(schoolData_complete["reading_score"]>70)]
passing_reading.reading_score.count()
```

```python
schoolData_complete["per_passing_read"]=(passing_reading.reading_score.count()/schoolData_complete.reading_score.count())
schoolData_complete.per_passing_read.head(1)
```

```python
#Create a dataframe to hold the above results
District_data = schoolData_complete[["school_count", "total_students", "district_budget",
                            "per_passing_math","per_passing_read","avg_math_score", "avg_reading_score", "overall_avg_score"]]
#Optional: give the displayed data cleaner formatting
District_data["district_budget"]=District_data["district_budget"].astype(float).map("${:,.2f}".format)
District_data["per_passing_math"]=District_data["per_passing_math"].astype(float).map('{:,.2%}'.format)
District_data["per_passing_read"]=District_data["per_passing_read"].astype(float).map('{:,.2%}'.format)

District_data.head(1)
```

```python
school_summary=school_data.sort_values("school_name")
school_summary

school_summary=school_summary.set_index("School ID")

school_summary["Stu_Budget"]=(school_summary["budget"]/school_summary["size"])
school_summary["Avg_Math"]=avg_math
school_summary["Avg_Read"]=avg_read
school_summary["Pass_Math"]=(Pass_Math/school_summary["size"])
school_summary["Pass_Read"]=(Pass_Read/school_summary["size"])
school_summary["Overall_Pass_Rate"]=(school_summary["Pass_Math"]+school_summary["Pass_Read"])/2
```
